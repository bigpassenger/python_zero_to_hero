```python
jobs = []

while True:
    user_input = str(input('Type add, show, edit, complete or exit: '))


    try:
        user_input, index = user_input.split(' ')
        index = int(index) - 1
    except:
        pass


    if user_input == "add":
        i = len(jobs) + 1
        user_job = str(input(f'Add task {i}: '))
        jobs.append(user_job)
        print(f'{user_job} is added')
    
    elif user_input == "show":
        j = 1
        for job in jobs:
            print(f"{j}. {job}")
            j = j + 1


    elif user_input == "edit":
        new_job = str(input('Add your new job: '))
        jobs[index] = new_job
        print('Your Task has just edited')


    elif user_input == "complete":
        job = jobs.pop(index)
        print(f'{job} has just deleted')


    elif user_input == "exit":
        break


    else:
        continue
```
